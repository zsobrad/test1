Load DexTools's ETH 24h Gainers

```python
import ccxt
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
import time
from IPython.display import clear_output, display

# Initialize exchanges with extended timeout
mexc = ccxt.mexc({'timeout': 30000})  # 30 seconds
kucoin = ccxt.kucoin()
xt = ccxt.xt()

# Function to fetch OHLCV data with retries
def fetch_ohlcv_with_retry(exchange, symbol, since=None, retries=3, delay=5):
    attempt = 0
    while attempt < retries:
        try:
            return fetch_ohlcv(exchange, symbol, since)
        except ccxt.RequestTimeout:
            attempt += 1
            if attempt >= retries:
                raise
            time.sleep(delay)
            print(f"Retrying... ({attempt}/{retries})")

# Function to fetch OHLCV data
def fetch_ohlcv(exchange, symbol, since=None):
    if since:
        since_timestamp = int(pd.Timestamp(since).timestamp() * 1000)  # Convert date to milliseconds
        return exchange.fetch_ohlcv(symbol, timeframe='1d', since=since_timestamp)
    else:
        return exchange.fetch_ohlcv(symbol, timeframe='1d')

# Fetch data for each token using retry logic
def fetch_token_data():
    stader_data = fetch_ohlcv_with_retry(mexc, 'SD/USDT', '2024-06-01T00:00:00Z')
    ton_data = fetch_ohlcv_with_retry(kucoin, 'TON/USDT', '2023-12-01T00:00:00Z')
    npc_data = fetch_ohlcv_with_retry(mexc, 'NPC/USDT', '2024-06-01T00:00:00Z')
    why_data = fetch_ohlcv_with_retry(xt, 'WHY/USDT', '2024-06-01T00:00:00Z')
    btc_data = fetch_ohlcv_with_retry(kucoin, 'BTC/USDT', '2023-12-01T00:00:00Z')  # Fetch BTC/USDT data
    return stader_data, ton_data, npc_data, why_data, btc_data

# Convert data to DataFrame for plotting
def data_to_df(data):
    df = pd.DataFrame(data, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
    df.set_index('timestamp', inplace=True)
    return df

# Calculate percentage change based on purchase price
def calculate_pnl(df, purchase_price):
    return (df['close'] / purchase_price - 1) * 100

# Special case for Stader PNL calculation
def calculate_stader_pnl(purchase_price, exit_price):
    return ((exit_price / purchase_price) - 1) * 100

# Plot data with Wintermute Trading style
def plot_data():
    stader_data, ton_data, npc_data, why_data, btc_data = fetch_token_data()
    
    df_stader = data_to_df(stader_data)
    df_ton = data_to_df(ton_data)
    df_npc = data_to_df(npc_data)
    df_why = data_to_df(why_data)
    df_btc = data_to_df(btc_data)
    
    # Calculate PNL percentages
    stader_pnl = calculate_stader_pnl(0.5955, 0.7986)  # Stader PNL based on buy and exit price
    df_ton['TON_pnl'] = calculate_pnl(df_ton, 2.14)
    df_npc['NPC_pnl'] = calculate_pnl(df_npc, 0.0174)
    df_why['WHY_pnl'] = calculate_pnl(df_why, 0.00000012)
    
    # Calculate BTC/USDT percentage change from December 2023
    df_btc['BTC_pnl'] = calculate_pnl(df_btc, df_btc['close'].iloc[0])
    
    # Combine all PNLs into a single DataFrame
    df_combined = pd.concat([pd.Series(stader_pnl, index=df_stader.index, name='SD_pnl'),
                             df_ton['TON_pnl'],
                             df_npc['NPC_pnl'],
                             df_why['WHY_pnl'],
                             df_btc['BTC_pnl']], axis=1)
    df_combined.columns = ['SD_pnl', 'TON_pnl', 'NPC_pnl', 'WHY_pnl', 'BTC_pnl']
    
    # Include Stader PNL in Portfolio Calculation
    df_combined['Portfolio_pnl'] = df_combined[['TON_pnl', 'NPC_pnl', 'WHY_pnl']].sum(axis=1) + stader_pnl
    
    # Calculate the maximum portfolio PNL
    max_portfolio_pnl = df_combined['Portfolio_pnl'].max()
    max_portfolio_pnl_date = df_combined['Portfolio_pnl'].idxmax()
    
    # Set Wintermute Trading style
    plt.style.use('dark_background')
    plt.figure(figsize=(20, 18))  # Adjusted height for 2x2 grid
    background_color = '#0c1c24'
    grid_color = '#1c303a'
    line_colors = ['#00d9e0', '#ff9800', '#00e676', '#f44336', '#9c27b0']
    
    plt.rcParams.update({
        'axes.facecolor': background_color,
        'axes.edgecolor': grid_color,
        'axes.labelcolor': 'white',
        'xtick.color': 'white',
        'ytick.color': 'white',
        'grid.color': grid_color,
        'grid.linestyle': '--',
        'text.color': 'white',
        'legend.frameon': False,
    })

    # Plot portfolio performance with BTC/USDT comparison
    plt.subplot(5, 1, 1)
    plt.plot(df_combined.index, df_combined['Portfolio_pnl'], label='My Main Portfolio', color=line_colors[2])
    plt.plot(df_combined.index, df_combined['BTC_pnl'], label='BTC/USDT', color=line_colors[1], linestyle='--')
    plt.title('Portfolio Performance vs BTC/USDT Over Time', fontsize=14)
    plt.xlabel('Date')
    plt.ylabel('Performance (%)')
    plt.axhline(0, color='white', linestyle='--', linewidth=0.5)
    plt.legend()

    # Annotate final percentage change in the right bottom corner
    portfolio_final = df_combined['Portfolio_pnl'].iloc[-1]
    btc_final = df_combined['BTC_pnl'].iloc[-1]
    plt.annotate(f'{portfolio_final:.2f}%', 
                 xy=(1, 0), 
                 xytext=(-10, 10), 
                 textcoords='offset points', 
                 ha='right', va='bottom',
                 fontsize=12, color=line_colors[0])
    plt.annotate(f'{btc_final:.2f}%', 
                 xy=(1, 0.1), 
                 xytext=(-10, 10), 
                 textcoords='offset points', 
                 ha='right', va='bottom',
                 fontsize=12, color=line_colors[1])

    # Annotate the maximum PNL value
    plt.annotate(f'Max: {max_portfolio_pnl:.2f}%', 
                 xy=(max_portfolio_pnl_date, max_portfolio_pnl), 
                 xytext=(10, 20), 
                 textcoords='offset points', 
                 arrowprops=dict(facecolor='white', arrowstyle='->'),
                 fontsize=12, color='white')

    # Add text next to the subplot (similar to other charts)
    plt.gca().text(1.05, 0.9, 
                   f'Portfolio PNL%: {portfolio_final:.2f}%\n'
                   f'BTC PNL%: {btc_final:.2f}%',
                   transform=plt.gca().transAxes, fontsize=10, verticalalignment='top', color=line_colors[2], bbox=dict(facecolor=background_color, alpha=0.8))

    # Plot SD/USDT performance
    plt.subplot(5, 2, 3)
    plt.plot(df_stader.index, df_stader['close'], label='SD/USDT', color=line_colors[2])
    plt.title('SD/USDT Performance Over Time', fontsize=12)
    plt.xlabel('Date')
    plt.ylabel('Price (USDT)')
    plt.axhline(0, color='white', linestyle='--', linewidth=0.5)
    plt.xlim(pd.Timestamp('2024-06-20'), pd.Timestamp('2024-07-11'))
    plt.legend()

    # Mark purchase and sale dates
    plt.axvline(pd.Timestamp('2024-06-23'), color='red', linestyle='--', linewidth=1, label='Bought on June 23, 2024')
    plt.axvline(pd.Timestamp('2024-07-10'), color='green', linestyle='--', linewidth=1, label='Sold on July 10, 2024')

    # Add text next to each subplot
    plt.gca().text(1.05, 0.9, 
                   f'Bought: 0.5955 USDT\n'
                   f'Sold: 0.7986 USDT\n'
                   f'PNL%: {stader_pnl:.2f}%',
                   transform=plt.gca().transAxes, fontsize=10, verticalalignment='top', color=line_colors[2], bbox=dict(facecolor=background_color, alpha=0.8))

    # Plot TON/USDT performance
    plt.subplot(5, 2, 4)
    plt.plot(df_ton.index, df_ton['close'], label='TON/USDT', color=line_colors[2])
    plt.title('TON/USDT Performance Over Time', fontsize=12)
    plt.xlabel('Date')
    plt.ylabel('Price (USDT)')
    plt.axhline(0, color='white', linestyle='--', linewidth=0.5)
    plt.legend()

    # Mark purchase date
    plt.axvline(pd.Timestamp('2023-12-29'), color='red', linestyle='--', linewidth=1, label='Bought on Dec 29, 2023')

    # Add text next to each subplot
    plt.gca().text(1.05, 0.9, 
                   f'Bought: 2.14 USDT\n'
                   f'Current: {df_ton["close"].iloc[-1]:.4f} USDT\n'
                   f'PNL%: {df_ton["TON_pnl"].iloc[-1]:.2f}%',
                   transform=plt.gca().transAxes, fontsize=10, verticalalignment='top', color=line_colors[2], bbox=dict(facecolor=background_color, alpha=0.8))

    # Plot NPC/USDT performance
    plt.subplot(5, 2, 5)
    plt.plot(df_npc.index, df_npc['close'], label='NPC/USDT', color=line_colors[2])
    plt.title('NPC/USDT Performance Over Time', fontsize=12)
    plt.xlabel('Date')
    plt.ylabel('Price (USDT)')
    plt.axhline(0, color='white', linestyle='--', linewidth=0.5)
    plt.legend()

    # Mark purchase date
    plt.axvline(pd.Timestamp('2024-06-27'), color='red', linestyle='--', linewidth=1, label='Bought on June 27, 2024')

    # Add text next to each subplot
    plt.gca().text(1.05, 0.9, 
                   f'Bought: 0.0174 USDT\n'
                   f'Current: {df_npc["close"].iloc[-1]:.4f} USDT\n'
                   f'PNL%: {df_npc["NPC_pnl"].iloc[-1]:.2f}%',
                   transform=plt.gca().transAxes, fontsize=10, verticalalignment='top', color=line_colors[2], bbox=dict(facecolor=background_color, alpha=0.8))

    # Plot WHY/USDT performance
    plt.subplot(5, 2, 6)
    plt.plot(df_why.index, df_why['close'], label='WHY/USDT', color=line_colors[2])
    plt.title('WHY/USDT Performance Over Time', fontsize=12)
    plt.xlabel('Date')
    plt.ylabel('Price (USDT)')
    plt.axhline(0, color='white', linestyle='--', linewidth=0.5)
    plt.legend()

    # Mark purchase date
    plt.axvline(pd.Timestamp('2024-06-27'), color='red', linestyle='--', linewidth=1, label='Bought on June 27, 2024')

    # Add text next to each subplot
    plt.gca().text(1.05, 0.9, 
                   f'Bought: 0.00000012 USDT\n'
                   f'Current: {df_why["close"].iloc[-1]:.10f} USDT\n'
                   f'PNL%: {df_why["WHY_pnl"].iloc[-1]:.2f}%',
                   transform=plt.gca().transAxes, fontsize=10, verticalalignment='top', color=line_colors[2], bbox=dict(facecolor=background_color, alpha=0.8))

    plt.tight_layout()
    plt.show()


# Continuous update loop
while True:
    clear_output(wait=True)
    plot_data()
    time.sleep(1)  # Refresh every second
```
